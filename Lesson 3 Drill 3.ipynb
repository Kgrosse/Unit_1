```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Admission_Predict.csv')
```

```python
#Choose one variable and plot that variable four different ways.
df.head(20)
```

```python
#Plot a variable four ways

#GRE Score Histogram
plt.hist(df['GRE Score'], alpha=.5, label='GRE Score')
plt.ylabel('Number of Applicants')
plt.xlabel('GRE Score')
plt.title('GRE Scores of Graduate School Applicants')
```

```python
#GRE Score Line Graph
plt.plot(df['GRE Score'], alpha=.5, label='GRE Score')
plt.ylabel('Number of Applicants')
plt.xlabel('GRE Score')
plt.title('GRE Scores of Graduate School Applicants')
```

```python
#GRE Score Boxplot
plt.boxplot(df['GRE Score'])
plt.ylabel('Number of Applicants')
plt.xlabel('GRE Score')
plt.title('GRE Scores of Graduate School Applicants')
```

```python
#GRE Score Barchart

#Group by number of applicants and calculate mean
y_group = df['GRE Score'].groupby(df['Serial No.']).mean()
x_axis = y_group.index
plt.bar(x_axis, y_group, alpha = 0.5)
plt.ylabel('Number of Applicants')
plt.xlabel('GRE Score')
plt.title('GRE Scores of Graduate School Applicants')
```

```python
#Choose two continuous variables, and plot them three different ways.

#Create a bar chart visualization of GRE Score and Chance of Admission
plt.bar(df['GRE Score'], df['Chance of Admit '])
plt.ylabel('Chance of Admission')
plt.xlabel('GRE Score')
plt.title('GRE Score vs. Chance of Admission')
```

```python
#Create a box plot visualization of GRE Score and Chance of Admission
plt.figure(figsize = (20,20))
ax = sns.boxplot(x='GRE Score',y='Chance of Admit ', data=df)  
ax.set(xlabel='Conducted Research', ylabel='GRE Score')
plt.ylabel('Chance of Admission')
plt.xlabel('GRE Score')
plt.title('GRE Score vs. Chance of Admission')
```

```python
#Create a scatter plot visualization of GRE Score and Chance of Admission
plt.scatter(df['GRE Score'], df['Chance of Admit '], alpha=.5, label='Research')  # alpha just controls the opacity
```

```python
#Choose one continuous variable and one categorical variable, and plot them six different ways.

#Create discrete variables
research = df[df['Research'] == 1]
no_research = df[df['Research']==0]

#First plot--Histogram
plt.hist(research['Research'], alpha=.5, label='Research')  # alpha just controls the opacity
plt.hist(no_research['Research'], color='blue', alpha=.5,label='No Research')
my_xticks = ["No Research", 'Research']
x_pos = list(range(len(my_xticks)))
plt.xticks(x_pos, my_xticks, fontsize = 10, rotation = 0
          )
plt.ylabel('Number of Applicants')
plt.title('The number of Graduate School Applicants who Conducted Research prior to \nApplying vs. Those who did not')

```

```python
#SecondPlot, overlapping histogram for GRE Scores for people who did and not conduct research

research_gre = research.loc[(research['Research']==1),'GRE Score']
noresearch_gre =  no_research.loc[(no_research['Research']==0),'GRE Score']

#Make overlapping histogram
plt.hist(research_gre, color='red',  alpha=.5,label='Conducted Reseach')  # alpha just controls the opacity
plt.hist(noresearch_gre, color='blue', alpha=.5,label='Did Not Conduct Research')
plt.xlabel('GRE Score')
plt.ylabel('Number of Applicants')
plt.legend(loc='upper right')
#plt.hist(mendinnertip, color='green', alpha=.5,label='Men at Dinner')  
#plt.hist(womendinnertip, color='yellow', alpha=.5,label='Women at Dinner')
```

```python
#Third plot - box plot

ax = sns.boxplot(x='Research',y='GRE Score',hue = 'Research',data=df)  
#sns.despine(offset=10, trim=True)
ax.set(xlabel='', ylabel='GRE Score')

#ay = sns.boxplot(x='Research',y='GRE Score',hue ='Research', data=no_research)  
#sns.despine(offset=10, trim=True)
#ax.set(xlabel='No Research', ylabel='GRE Score')

#use subplot to
```

```python
plt.scatter(research['GRE Score'], research['Chance of Admit '], alpha = 0.5, label = "Research")
plt.scatter(no_research['GRE Score'], no_research['Chance of Admit '], alpha = 0.5, label = "No Research")
plt.xlabel('GRE Score')
plt.ylabel('Chance of Admission')
plt.legend(loc='lower right')
```

```python
#Fifth Plot - Histogram subplots
sns.set(style="ticks")  #Setting the overall aesthetic
# Tell seaborn about the structure of our data.
g = sns.FacetGrid(df, row="Research")
# Name the plot type and the variable to be plotted using the structure.
g.map(plt.hist, "Chance of Admit ", color="steelblue",  lw=0)
# Moving the plots apart to make room for our titles.
plt.subplots_adjust(top=0.9)
# Making a more informative axis name.
g.set_axis_labels('GRE Score')
plt.suptitle('Plot 1: GRE Score by Research')
# Removing excess lines around the plot.
sns.despine(trim=True)
plt.show()
```

```python
#Sixth Plot - overlapping bar charts
plt.bar(df['GRE Score'], df['Chance of Admit '], df['Research']==1, alpha = 0.5, label ='Research')
plt.bar(df['GRE Score'], df['Chance of Admit '], df['Research']==0, alpha = 0.5,label = "No Research")
plt.legend(loc='lower right')
plt.ylabel('Chance of Admission')
plt.xlabel('GRE Score')
```
